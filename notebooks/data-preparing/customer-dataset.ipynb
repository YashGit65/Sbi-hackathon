```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
import os
```

```python
NUM_CUSTOMERS = 1000

professions = [
    "Software Engineer", "Doctor", "Teacher", "Business Owner",
    "Government Employee", "Student", "Lawyer",
    "Farmer", "Marketing Manager", "Chartered Accountant"
]

locations = [
    "Mumbai", "Delhi", "Bangalore", "Chennai",
    "Hyderabad", "Pune", "Lucknow", "Jaipur",
    "Agra", "Varanasi"
]
```

```python
risk_profiles = ["Conservative", "Moderate", "Aggressive"]
customers = []
```

```python
for i in range(1, NUM_CUSTOMERS + 1):

    customer_id = f"C{i:04d}"

    age = random.randint(21, 65)

    profession = random.choice(professions)

    income = random.randint(20000, 250000)

    marital_status = random.choice(["Single", "Married"])

    location = random.choice(locations)

    

    customers.append([
        customer_id,
        age,
        profession,
        income,
        marital_status,
        location
    ])
```

```python
customer_df = pd.DataFrame(
    customers,
    columns=[
        "customer_id",
        "age",
        "profession",
        "income",
        "marital_status",
        "location",
        "risk_profile",
        "existing_investment"
    ]
)
```

```python
categories = [
    "Food",
    "Shopping",
    "Travel",
    "Entertainment",
    "Utilities",
    "Healthcare",
    "Education",
    "Investment"
]
```

```python
transactions = []

start_date = datetime(2025, 1, 1)

for customer_id in customer_df["customer_id"]:

    num_transactions = random.randint(10, 30)

    for _ in range(num_transactions):

        category = random.choice(categories)

        amount = random.randint(100, 15000)

        random_days = random.randint(0, 365)

        txn_date = start_date + timedelta(days=random_days)

        transactions.append([
            customer_id,
            txn_date.strftime("%Y-%m-%d"),
            category,
            amount
        ])
```

```python
transaction_df = pd.DataFrame(
    transactions,
    columns=[
        "customer_id",
        "transaction_date",
        "category",
        "amount"
    ]
)
```

```python
transaction_path = "data/raw"
transaction_df.to_csv(transaction_path, index=False)
```
