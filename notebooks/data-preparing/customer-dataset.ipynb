```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
import os
from faker import Faker
```

```python
NUM_CUSTOMERS = 1000

professions = [
    "Software Engineer", "Doctor", "Teacher", "Business Owner",
    "Government Employee", "Student", "Lawyer",
    "Farmer", "Marketing Manager", "Chartered Accountant"
]

locations = [
    "Mumbai", "Delhi", "Bangalore", "Chennai",
    "Hyderabad", "Pune", "Lucknow", "Jaipur",
    "Agra", "Varanasi"
]
```

```python
risk_profiles = ["Conservative", "Moderate", "Aggressive"]
fake = Faker("en_IN")  # Indian names
customers = []
```

```python
for i in range(1, NUM_CUSTOMERS + 1):

    customer_id = f"C{i:04d}"
    
    name = fake.name()
    
    phone_number = f"{random.choice(['6','7','8','9'])}{random.randint(100000000, 999999999)}"

    age = random.randint(21, 65)

    profession = random.choice(professions)

    income = random.randint(20000, 250000)

    marital_status = random.choice(["Single", "Married"])

    location = random.choice(locations)

    

    customers.append([
        customer_id,
        name,
        phone_number,
        age,
        profession,
        income,
        marital_status,
        location
    ])
```

```python
customer_df = pd.DataFrame(
    customers,
    columns=[
        "customer_id",
        "name",
        "phone_number",
        "age",
        "profession",
        "income",
        "marital_status",
        "location"
    ]
)
```

```python
categories = [
    "Food",
    "Shopping",
    "Travel",
    "Entertainment",
    "Utilities",
    "Healthcare",
    "Education",
    "Investment"
]
```

```python
transactions = []

start_date = datetime(2025, 1, 1)

for customer_id in customer_df["customer_id"]:

    num_transactions = random.randint(10, 30)

    for _ in range(num_transactions):

        category = random.choice(categories)

        amount = random.randint(100, 15000)

        random_days = random.randint(0, 365)

        txn_date = start_date + timedelta(days=random_days)

        transactions.append([
            customer_id,
            txn_date.strftime("%Y-%m-%d"),
            category,
            amount
        ])
```

```python
transaction_df = pd.DataFrame(
    transactions,
    columns=[
        "customer_id",
        "transaction_date",
        "category",
        "amount"
    ]
)
```

```python
transaction_df.to_csv(raw_data_path / "transactions.csv", index=False)
```

```python


def recommendation_agent(risk, goals, income, savings_rate):

    recommendations = []

    # Determine investment capacity
    if savings_rate >= 30:
        invest_ratio = 0.8

    elif savings_rate >= 15:
        invest_ratio = 0.6

    else:
        invest_ratio = 0.3

    investable_amount = income * (savings_rate / 100) * invest_ratio

    # Emergency Fund
    if "Build Emergency Fund" in goals:

        recommendations.append({
            "goal": "Build Emergency Fund",
            "product": "FD / Liquid Fund",
            "amount": round(investable_amount * 0.3, 2)
        })

    # Retirement
    if "Retirement Planning" in goals:

        if risk == "Aggressive":
            product = "Equity SIP"

        elif risk == "Moderate":
            product = "Balanced SIP"

        else:
            product = "PPF / Debt Fund"

        recommendations.append({
            "goal": "Retirement Planning",
            "product": product,
            "amount": round(investable_amount * 0.4, 2)
        })

    # Child Education
    if "Child Education Planning" in goals:

        recommendations.append({
            "goal": "Child Education Planning",
            "product": "Hybrid Fund",
            "amount": round(investable_amount * 0.3, 2)
        })

    # Insurance
    if income > 50000:
        insurance = "Term + Health Insurance"
    else:
        insurance = "Health Insurance"

    return {
        "risk_profile": risk,
        "savings_rate": round(savings_rate, 2),
        "investable_amount": round(investable_amount, 2),
        "insurance": insurance,
        "recommendations": recommendations
    }
```
